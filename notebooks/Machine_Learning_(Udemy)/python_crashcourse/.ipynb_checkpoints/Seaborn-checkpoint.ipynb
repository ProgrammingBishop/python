```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
# Distribution Plots
tips = sns.load_dataset( 'tips' )
tips.head()
```

```python
# KDE = Kernal Desnity Estimation
sns.distplot(tips['total_bill'], 
             kde = False,
             bins = 40 )
```

```python
sns.jointplot(x = 'total_bill', 
              y = 'tip', 
              data = tips, 
              kind = 'hex' )
```

```python
sns.jointplot(x = 'total_bill', 
              y = 'tip', 
              data = tips, 
              kind = 'reg' )
```

```python
sns.jointplot(x = 'total_bill', 
              y = 'tip', 
              data = tips, 
              kind = 'kde' )
```

```python
sns.jointplot(x = 'total_bill', 
              y = 'tip', 
              data = tips, 
              kind = 'scatter' )
```

```python
# hue will color datapoints based on categorical classifications
sns.pairplot( tips, hue = 'sex', palette = 'coolwarm' )
```

```python
# Rug Plot draws dashed line for every point in univariate distribution 
sns.rugplot( tips['total_bill'] )
```

```python
# Categorical Plots
```

```python
sns.barplot(x = 'sex', 
            y = 'total_bill', 
            data = tips, 
            estimator = np.std )
```

```python
sns.countplot( x = 'sex', data = tips )
```

```python
sns.boxplot(x = 'day', 
            y = 'total_bill', 
            data = tips, 
            hue = 'smoker' )
```

```python
sns.stripplot(x = 'day', 
              y = 'total_bill', 
              data = tips, 
              jitter = True, 
              hue = 'sex', 
              dodge = True )
```

```python
sns.swarmplot(x = 'day', 
              y = 'total_bill', 
              data = tips, 
              hue = 'sex', 
              dodge = True )
```

```python
# Matrix Plots
```

```python
flights = sns.load_dataset( 'flights' )
flights.head()
```

```python
tips.head()
```

```python
tc = tips.corr()
```

```python
sns.heatmap( tc, annot = True )
```

```python
# Transform data into matrix form
fp = flights.pivot_table( index = 'month', columns = 'year', values = 'passengers' )
```

```python
sns.heatmap(fp, 
            cmap = 'magma', 
            linecolor = 'white', 
            linewidth = 0.05 )
```

```python
sns.clustermap(fp, 
               standard_scale = 1, 
               linecolor = 'white', 
               linewidth = 0.05 )
```

```python
# Grids 
```

```python
iris = sns.load_dataset( 'iris' )
iris.head()
```

```python
iris[ 'species' ].unique()
```

```python
# PairGrid allows autonomy on what plots are plotted where
g = sns.PairGrid( iris )
g.map_diag( sns.distplot )
g.map_upper( plt.scatter )
g.map_lower( sns.kdeplot )
```

```python
g = sns.FacetGrid( data = tips, col = 'time', row = 'smoker' )
g.map( plt.scatter, 'total_bill', 'tip' )
```

```python
# Regression Plots
```

```python
sns.lmplot(x = 'total_bill', 
           y = 'tip', 
           data = tips, 
           hue = 'sex', 
           markers = [ 'o', 'v' ], 
           # Call matplotlib styles under the hood
           scatter_kws = { 's' : 100 } )
```

```python
sns.lmplot(x = 'total_bill', 
           y = 'tip', 
           data = tips, 
           col = 'day', 
           hue = 'sex', 
           aspect = 0.6, 
           size = 8 )
```

```python
# Seaborn Styles
sns.set_style( 'darkgrid' )
sns.set_context( 'poster', font_scale = 1 )
plt.figure( figsize = (12, 6) )

sns.countplot( x = 'sex', data = tips )
sns.despine()
```
