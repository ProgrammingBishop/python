```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Allows you to see plots inline in Jupyter Notebook
%matplotlib inline
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
# Functional Method
plt.plot( x, y, 'r-' )
plt.xlabel( 'X Lable' )
plt.ylabel( 'Y Label' )
plt.title( 'Title' )
```

```python
# plt.subplot( num_rows, num_col, fig_pos )
plt.subplot( 1, 2, 1 )
plt.plot( x, y, 'r-' )

plt.subplot( 1, 2, 2 )
plt.plot( y, x, 'b' )
```

```python
# Object-oriented method

# Create a figure object to plot onto
fig = plt.figure()

# add_axes( [ left, bottom, width, height ] )
axes = fig.add_axes( [ 0.1, 0.1, 0.8, 0.8 ] )
axes.plot( x, y )
axes.set_xlabel( 'X Lable' )
axes.set_ylabel( 'Y Label' )
axes.set_title( 'Title' )
```

```python
fig = plt.figure()

axes1 = fig.add_axes( [0.1, 0.1, 0.8, 0.8] )

# Set a subplot within another plot (based on percent in parent plot)
axes2 = fig.add_axes( [0.2, 0.5, 0.4, 0.3] )
```

```python
# Axes is an array of matplotlib objects (it is iterable)
fig, axes = plt.subplots( nrows = 1, ncols = 2 )

for current_ax in axes:
    current_ax.plot( x, y )
```

```python
fig, axes = plt.subplots( nrows = 1, ncols = 2 )

axes[0].plot( x, y )
axes[0].set_title( 'First Plot' )

axes[1].plot( y, x )
axes[1].set_title( 'Second Plot' )

# Fix overlapping
plt.tight_layout()
```

```python
# Figure Size and DPI
```

```python
# figsize( fig_width, fig_height ) in inches
fig, axes = plt.subplots( nrows = 2, ncols = 1, figsize = ( 8, 4 ), dpi = 100 )
axes[0].plot( x, y )
axes[1].plot( y, x )
plt.tight_layout()
```

```python
# Save a figure
fig.savefig( 'filename.jpg', dpi = 200 )
```

```python
# Legends
fig = plt.figure()
ax = fig.add_axes( [ 0, 0, 1, 1 ] )
ax.plot( x, y ** 2, label = 'Y Squared' )
ax.plot( y, x ** 3, label = 'X Cubed' )

# You can also set by tuple location ( 0.1, 0.1 )
ax.legend( loc = 0 )
```

```python
# Set Colors
fig = plt.figure()
ax = fig.add_axes( [0, 0, 1, 1] )
ax.plot( x, y, 
        color = '#ff8c00', 
        lw = 1, 
        #alpha = 0.5, 
        ls = '-', 
        marker = 'o', 
        markersize = 5,
        markerfacecolor = 'orange', 
        markeredgewidth = 2,
        markeredgecolor = 'orange' )

# Control Axes
ax.set_xlim( [ 0, 10 ] )
ax.set_ylim( [ 0, 30 ] )
```
